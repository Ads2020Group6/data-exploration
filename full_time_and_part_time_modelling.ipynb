```python
# Discrete value correlation coefficient
import math
def getEntropy(s):
    if not isinstance(s,pd.core.series.Series):
        s=pd.Series(s)
    prt_ary=pd.Series.groupby(s,by=s).count().values/float(len(s))
    return -(np.log2(prt_ary)*prt_ary).sum()
def getCondEntropy(s1,s2):
    d=dict()
    for i in list(range(len(s1))):
        d[s1[i]]=d.get(s1[i],[])+[s2[i]]
    return sum([getEntropy(d[k])*len(d[k])/float(len(s1)) for k in d])

def getEntropyGain(s1,s2):
    return getEntropy(s2)-getCondEntropy(s1,s2)

# (0,1),1 positive correlation, 0 uncorrelated
def getDiscreteCorr(s1,s2): 
    return getEntropyGain(s1,s2)/math.sqrt(getEntropy(s1)*getEntropy(s2))

print(getDiscreteCorr(df5['Age group'],df5['Type']))
print(getDiscreteCorr(df5['Age group'],df5['Value']))
print(getDiscreteCorr(df5['Type'],df5['Value']))
```

```python

from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.preprocessing import Normalizer
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
# Feature processing
def processing(ty,ag,lower_d,ld_n,df5):
    label = df5["Value"]
    df5 = df5.drop("Value", axis=1)
    # Digitize
    scaler_lst = [ty,ag]
    column_lst = ['Type', 'Age group']
    for i in range(len(scaler_lst)):
        if not scaler_lst[i]:
            if column_lst[i] == "Type" :
                df5[column_lst[i]] = [map_type(s) for s in df5["Type"].values]
            else:
                df5[column_lst[i]] = [map_age(s) for s in df5["Age group"].values]
            df5[column_lst[i]] = MinMaxScaler().fit_transform(df5[column_lst[i]].values.reshape(-1, 1)).reshape(1, -1)[0]
        else:
            df5 = pd.get_dummies(df5, columns=[column_lst[i]])
    if lower_d:
        return PCA(n_components=ld_n).fit_transform(df5.values), label
    return df5,label

d=dict([("Full-time",0),("Part-time",1)])
def map_type(s):
    return d.get(s,0)

d2=dict([("18 to 21 years",0),("22 to 29 years",1),("30 to 39 years",2),("40 to 49 years",3)])
def map_age(s):
    return d2.get(s,0)

def regr_test(features,label):
#     print("X",features)
#     print("Y",label)
    from sklearn.linear_model import LinearRegression,Ridge,Lasso
    regr=LinearRegression()
#     regr=Ridge(alpha=0)  
    # Lasso
#     regr = Lasso(alpha=1)

    regr.fit(features.values,label.values)
    Y_pred=regr.predict(features.values)
    print("Coef:",regr.coef_) 
    
    from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
    print("MSE:",mean_squared_error(label.values,Y_pred))
    print("MAE:",mean_absolute_error(label.values,Y_pred))
    print("R2:",r2_score(label.values,Y_pred))
    
features,label=processing(False,False,False,1,df5)
regr_test(features, label)
```

```python
def modeling(features,label):
    from sklearn.model_selection import train_test_split
    f_v = features.values
    f_names = features.columns.values
    l_v = label.values
    X_tt, X_validation, Y_tt, Y_validation = train_test_split(f_v, l_v, test_size=0.2)
    X_train, X_test, Y_train, Y_test = train_test_split(X_tt, Y_tt, test_size=0.25)
    # print(len(X_train),len(X_validation),len(X_test))

    from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
    from sklearn.tree import DecisionTreeRegressor
    from sklearn.svm import LinearSVC
    from sklearn.ensemble import RandomForestRegressor
    from sklearn.ensemble import AdaBoostRegressor
    from sklearn.ensemble import GradientBoostingRegressor
  
    models = []

    models.append(("DecisionTreeGini", DecisionTreeRegressor()))
    models.append(("SVM", LinearSVC(C=1000)))
    models.append(("RandomForest", RandomForestRegressor(n_estimators=11, max_features=None)))
    models.append(("Adaboost", AdaBoostRegressor(n_estimators=100)))
    models.append(("GBDT",GradientBoostingRegressor(max_depth=6,n_estimators=100)))

    for clf_name, clf in models:
        clf.fit(X_train, Y_train.astype('int'))
        xy_lst = [(X_train, Y_train), (X_validation, Y_validation), (X_test, Y_test)]
        for i in range(len(xy_lst)):
            X_part = xy_lst[i][0]
            Y_part = xy_lst[i][1]
            Y_pred = clf.predict(X_part)
            print(i)
            print(clf_name,"MSE:",mean_squared_error(Y_part,Y_pred))
            print(clf_name,"MAE:",mean_absolute_error(Y_part,Y_pred))
            print(clf_name,"R2:",r2_score(Y_part,Y_pred))
            
features,label=processing(False,False,False,1,df5)
# print(features,label)
modeling(features, label)
```
